```python
# Maskeyi RGB olarak oku
mask_bgr = cv2.imread(sample_mask_path)
mask_rgb = cv2.cvtColor(mask_bgr, cv2.COLOR_BGR2RGB)

print(f"Maske shape: {mask_rgb.shape}")

# Benzersiz renkleri bul
unique_colors = np.unique(mask_rgb.reshape(-1, 3), axis=0)
print(f"Maskedeki renkler:")
for i, color in enumerate(unique_colors):
    print(f"  Renk {i}: RGB{color}")

# Aynı maskeyi aynen görselleştir
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(mask_rgb)
plt.title('Orijinal Maske')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mask_rgb)  # Aynen aynısı
plt.title('Aynı Maske (Kontrol)')
plt.axis('off')

plt.show()

# Her renk için pixel sayısı
for i, color in enumerate(unique_colors):
    # Bu renkteki pixelleri bul
    mask_this_color = np.all(mask_rgb == color, axis=2)
    pixel_count = np.sum(mask_this_color)
    percentage = (pixel_count / mask_rgb.size) * 100
    print(f"RGB{color}: {pixel_count} pixel ({percentage:.2f}%)")
```
